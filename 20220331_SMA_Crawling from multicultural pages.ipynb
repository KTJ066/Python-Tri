```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python
url = "https://www.officialcharts.com/charts/dvd-chart/"
html = req.urlopen(url)
doc = BeautifulSoup(html, "html.parser")
```

```python
position = [p.text for p in doc.find_all(class_="position")]
print(position)
```

```python
## 이번에 크롤링해온 사이트 url을보니 영화차트를 주차별로 구분하고 url에 그 시작을 일요일의 날짜로 하고 있는걸 알았으니 url에 규칙을 넣어서 하나만으로 여러개 크롤링 해보자
```

```python
time = pd.date_range(start="2022-01-02", periods=5, freq="w-Sun") ##periods는 몇번의 기간동안 freqency 는 w : 주차별로, Sun : 일요일마다
time
```

```python
time[0]
```

```python
time[0].strftime("%Y,%m,%d") ## strftime 은 string from time : time에서 가져온 자료를 string으로 바꿔라
```

```python
time[0].strftime("%Y%m%d")
```

```python
## 오늘 수업은 이거 보여주려 한 것
dates = []
positions =[]
titles = []
studios = []

for t in time:
    x = t.strftime("%Y%m%d")
    url = "https://www.officialcharts.com/charts/dvd-chart/"+ x + "/141/"
    html = req.urlopen(url)
    doc = BeautifulSoup(html, "html.parser")
    
    d = [t for n in range(0,100)] ##date를 100번 넣어라
    dates.extend(d)
    
    p = [doc.find_all(class_="position")[n].text for n in range(0,100)]
    positions.extend(p)
    
    t = [doc.find_all(class_="title")[n].text.strip() for n in range(0,100)]
    titles.extend(t)
    
    s =[doc.find_all(class_="label-cat")[n].text for n in range(0,100)]
    studios.extend(s)
    
    
```

```python
dvd_rankings = pd.DataFrame({"Date": dates,
                            "Position": positions,
                            "Title": titles,
                            "Studio": studios})
dvd_rankings
```
