```python
import urllib.request as req
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
from selenium.webdriver.chrome.service import Service
```

```python
driver = webdriver.Chrome('/Users/jonghyunkim/repo/Tristan/chromedriver')
##window에서는 webdriver.Chrome(service=Service("chromedriver.exe")) - 내가 파이썬 저장하는 폴더에 크롬 드라이버 있을 때
```

```python
url = "https://www.edmunds.com/tesla/model-3/2021/consumer-reviews/?pagesize=50"
driver.get(url)
```

```python
doc = BeautifulSoup(driver.page_source, "html.parser")
print(doc.prettify())
```
