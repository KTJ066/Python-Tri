```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.common.keys import Keys
from time import sleep
```

```python
urls = []
for i in range (1, 5):
    url = "https://www.webmd.com/drugs/drugreview-180368-pfizer-biontech-covid-19-vaccine-pf-eua-intramuscular?conditionid=&sortval=1&pagenumber=" + str(i)
    urls.append(url)
```

```python
text_array = []
driver = webdriver.Chrome('/Users/jonghyunkim/repo/Tristan/chromedriver')
for u in urls:
    driver.get(u)
    sleep(1)
    page = BeautifulSoup(driver.page_source, "html.parser")
    text = page.find_all(class_="description-text")
    for n in range(0, len(text)):
        text_array.append(text[n].text.strip().replace("Read More Read Less", "").replace("\n",""))
    sleep(2)
driver.quit()
pd.DataFrame({"Review":text_array})
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
import nltk
nltk.download("punkt")
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
review_word_list = []
for n in text_array:
    text_t = sent_tokenize(n)
    review_word_list.append(text_t)

review_word_list
```

```python
from nltk.corpus import stopwords
nltk_stopwords = set(stopwords.words("english"))
```

```python
stem_review = []
for sl in review_word_list:
    stem_sent = []
    for s in sl:
        word_t = word_tokenize(s)
        word_stem = [st.lower() for st in word_t if st.isalpha()]
        word_stem2 = [st2 for st2 in word_stem if not st2 in nltk_stopwords]
        word_stem3 = [ps.stem(st3) for st3 in word_stem2]
        stem_sent.append(word_stem3)
    stem_review.append(stem_sent)
stem_review
```
