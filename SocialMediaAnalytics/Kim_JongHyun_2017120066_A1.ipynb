```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python
url = "https://money.cnn.com/data/markets/"
html = req.urlopen(url)
doc = BeautifulSoup(html, "html.parser")
```

```python
print(doc.prettify())
```

```python
stock = doc.find(class_="module-body wsod most-popular-stocks")
stock.prettify()
```

```python
stock_names = stock.find_all(class_="column stock-name")
stock_names
```

```python
for s in stock_names:
    print(s.string)
```

```python
names = [s.string for s in stock_names]
names
```

```python
stock_prices = stock.find_all(class_="column stock-price")
stock_prices
```

```python
for p in stock_prices:
    print(p.string)
```

```python
prices = [p.string for p in stock_prices]
prices
```

```python
stock_price_changes = stock.find_all(class_ = "column stock-change")
stock_price_changes
```

```python
for c in stock_price_changes:
    print(c.string)
```

```python
changes = [c.string for c in stock_price_changes]
changes
```

```python
import pandas as pd
```

```python
stock_table = pd.DataFrame({"Name": names, "Price": prices, "Change": changes})
stock_table
```

```python
stock_table.to_excel("popular_stocks_Kim_JongHyun_2017120066.xlsx")
```
