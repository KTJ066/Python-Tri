```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python
url1 = "https://www.amazon.com/SAMSUNG-Factory-Unlocked-Smartphone-Foldable/product-reviews/B097CNBDX2/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
header = "Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36"
url2 = req.Request(url1, headers = {"User-Agent" : header})

html = req.urlopen(url2)
doc = BeautifulSoup(html, "html.parser")
```

```python
print(doc.prettify())
```

```python
reviews = doc.find(id="cm_cr-review_list")
reviews
```

```python
reviewer = []
title = []
review = []
```

```python
n =[reviews.find_all(attrs={"data-hook":"genome-widget"})[n].text for n in range(0,10)]
reviewer.extend(n)
reviewer
```

```python
t =[reviews.find_all(attrs={"data-hook":"review-title"})[n].text.strip() for n in range(0,10)]
title.extend(t)
title
```

```python
b =[reviews.find_all(attrs={"data-hook":"review-body"})[n].text.strip() for n in range(0,10)]
review.extend(b)
review
```

```python
import pandas as pd
galaxy = {"Reviewer": reviewer, "Title": title, "Review": review}
Galaxy_review = pd.DataFrame(galaxy)
Galaxy_review
```

```python
Top_GR = Galaxy_review.head()
Top_GR
```

```python
Top_GR.to_excel("Galaxy_Fold_Amazon_Reviews.xlsx")
```
