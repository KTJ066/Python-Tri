```python
from selenium.webdriver.common.keys import Keys
```

```python
for n in range(0,20):
    driver.find_element(By.TAG_NAME, "body").send_keys(Keys.PAGE_DOWN)
```

```python
from time import sleep
```

```python
reviewers = []
reviews = []

for l in links:
    l.click()
    sleep(2)
    
    main_window = driver.current_window_handle
    
    driver.switch_to.window(driver.window_handles[1])
    
    doc = BeautifulSoup(driver.page_source, "html.parser")
    
    for t in doc.find_all(class_="oc-reviews-42b9159d"):
        t.decompose()
    
    end = len(doc.find_all(class_="oc-reviews-954a6007"))
    
    r1 = [doc.find_all(class_="oc-reviews-954a6007")[n].text for n in range(0,end)]
    reviewers.extend(r1)
    
    r2 = [doc.find_all(class_="reviewBodyContainer oc-reviews-8107696f")[n].text for n in range(0,end)]
    reviews.extend(r2)
    
    driver.close()
    driver.switch_to.window(main_window)
    sleep(2)

pd.DataFrame({"Reviewer": reviewers, "Review" : reviews})
```
