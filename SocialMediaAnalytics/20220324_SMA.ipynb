```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python
url = "https://forecast.weather.gov/MapClick.php?lat=40.93377000000004&lon=-73.89567999999997#.Yju1ri_kHgs"
html = req.urlopen(url)
doc = BeautifulSoup(html, "html.parser")
```

```python
print(doc.prettify())
```

```python
forcast = doc.find(id = 'seven-day-forecast-container')
```

```python
forcast.prettify()
```

```python
period_names = forcast.find_all(class_= 'period-name')
period_names
```

```python
for p in period_names:
    print(p.text) ##p.string 해서 안나오면 p.text 해보자
```

```python
[p.text for p in period_names]
```

```python
names = [p.get_text(separator=" ") for p in period_names] ##get_text(separator=" ")를 함으로서 period names 안에 있는 text를 가져올 때 2개를 가져올 때는 띄어쓰기를 해준다.
names
```

```python
short_descs = forcast.find_all(class_="short-desc")
short_descs
```

```python
short = [s.get_text(separator=" ") for s in short_descs]
short
```

```python
forcast.find_all(class_="temp temp-low") ## 9개가 필요한데 5개가 들어온다. 소스를 보면 temp temp-high로 분류된 애들도 잇어서 그렇다
```

```python
temps_low_high = forcast.find_all(class_="temp") ##temp temp-low 와 temp temp-high는 중간에 띄어쓰기로 나뉘어져 있는데, 이건 두개의 파트를 사용하고 있다는 것이다.
## 두개의 파트로 나뉘어져 있는 것을 가져올 때 뒤에 파트가 상이한 것들을 모두 가져오고 싶으면 앞에 파트만 쳐서 가져오면 된다.
temps_low_high
```

```python
temps = [t.get_text() for t in temps_low_high]
```

```python
temps
```

```python
icons = forcast.find_all(class_="forecast-icon")
```

```python
icons[0].attrs ##이제 여기서 가져오고 싶은 것은 alt에 있는 거나 title에 있는거다
```

```python
icons[0].attrs["title"]
```

```python
icons[0]["title"] ##attrs가 있는 코드와 같은 역할
```

```python
longs = [i["title"] for i in icons]
longs
```

```python
import pandas as pd
```

```python
forcast_table = pd.DataFrame({"Name": names, "Short Description": short, "Temperature": temps, "Full Description": longs}) ##DataFrame은 테이블을 만들라는 얘기다 pd도 테이블 만드는 애다
forcast_table
```

```python
forcast_table.to_csv("forcast.csv") ##따옴표 안에 있는 이름으로 컴퓨터에 파일이 저장됨 to_csv 는 csv 파일로 저장하라는거
```
