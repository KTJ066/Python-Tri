```python
time = pd.date_range(start="2022-01-02", periods=5, freq="w-Sun") ##periods   freqency  w : , Sun : 
time
```

```python
time[0]
```

```python
time[0].strftime("%Y,%m,%d") ## strftime  string from time : time   string 
```

```python
time[0].strftime("%Y%m%d")
```

```python
##      
dates = []
positions =[]
titles = []
studios = []

for t in time:
    x = t.strftime("%Y%m%d")
    url = "https://www.officialcharts.com/charts/dvd-chart/"+ x + "/141/"
    html = req.urlopen(url)
    doc = BeautifulSoup(html, "html.parser")
    
    d = [t for n in range(0,100)] ##date 100 
    dates.extend(d)
    
    p = [doc.find_all(class_="position")[n].text for n in range(0,100)]
    positions.extend(p)
    
    t = [doc.find_all(class_="title")[n].text.strip() for n in range(0,100)]
    titles.extend(t)
    
    s =[doc.find_all(class_="label-cat")[n].text for n in range(0,100)]
    studios.extend(s)
    
    
```

```python
dvd_rankings = pd.DataFrame({"Date": dates,
                            "Position": positions,
                            "Title": titles,
                            "Studio": studios})
dvd_rankings
```
