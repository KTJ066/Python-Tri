```python
pages = range(1,8)
list(pages)
```

```python
reviewers = []
reviews = []


for p in pages:
    url = "https://www.opentable.com/r/the-woo-new-york?page=" + str(p) ## p 는 int 자료형이라서 str로 바꿔주면서 더하기를 해야 같이 텍스트로 인식된다
    html = req.urlopen(url)
    doc = BeautifulSoup(html, "html.parser")
    
    for t in doc.find_all(class_='oc-reviews-42b9159d'):
        t.decompose()
    
    end = len(doc.find_all(class_='oc-reviews-954a6007'))
    
    r1 = [doc.find_all(class_='oc-reviews-954a6007')[n].text for n in range(0, end)]
    reviewers.extend(r1)
    
    r2 = [doc.find_all(class_="reviewBodyContainer oc-reviews-8107696f")[n].text for n in range(0, end)]
    reviews.extend(r2)
    
pd.DataFrame({"Reveiwer": reviewers, "Review": reviews})
```
