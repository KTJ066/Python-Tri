```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.common.keys import Keys
from time import sleep
```

```python
urls = []
for i in range (1, 5):
    url = "https://www.webmd.com/drugs/drugreview-180368-pfizer-biontech-covid-19-vaccine-pf-eua-intramuscular?conditionid=&sortval=1&pagenumber=" + str(i)
    urls.append(url)
```

```python
text_array = []
driver = webdriver.Chrome('/Users/jonghyunkim/repo/Tristan/chromedriver')
for u in urls:
    driver.get(u)
    sleep(1)
    page = BeautifulSoup(driver.page_source, "html.parser")
    text = page.find_all(class_="description-text")
    for n in range(0, len(text)):
        text_array.append(text[n].text.strip().replace("Read More Read Less", "").replace("\n",""))
    sleep(2)
driver.quit()
pd.DataFrame({"Review":text_array})
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction import text
```

```python
sklearn_stopwords = text.ENGLISH_STOP_WORDS
```

```python
cv = CountVectorizer(lowercase=True, stop_words=sklearn_stopwords)
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
import nltk
nltk.download("punkt")
```

```python
review_text = []
for r in text_array:
    sent = sent_tokenize(r)
    for s in sent:
        review_text.append(s)

review_text
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
sentiment_list = []
for st in review_text:
    sentiment = sid.polarity_scores(st)
    score = sentiment['compound']
    sentiment_list.append(score)
sentiment_list
```

```python
abnormal = list(zip(review_text, sentiment_list))
abnormal
```

```python
sentiment_review = []
for a in abnormal:
    if (a[1] != 0.0):
        sentiment_review.append(a[0])
sentiment_review
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
new_sentiment_list = []
for re in sentiment_review:
    n_sentiment = sid.polarity_scores(re)
    n_score = n_sentiment['compound']
    new_sentiment_list.append(n_score)
new_sentiment_list
```

```python
cv_reviews = cv.fit_transform(sentiment_review)
```
