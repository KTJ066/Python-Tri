```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python
url1 = "https://www.ultimatemovierankings.com/top-250-movies/"
header = "Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36" ## 2. 이걸 추가해줌으로써 legitamate 하게 만들 수 있다
url2 = req.Request(url1, headers = {"User-Agent" : header}) ##2-2 url2를 보낼 때 url1과 header 정보를 같이 보내게 해준다.
##가끔 0324 수업까지의 url보내는 걸로는 크롤링 시에 막히는 경우가 있다. 그럴 땐 header 정보를 추가해서 url2처럼 만들어주면 해결된다.

html = req.urlopen(url2)  
doc = BeautifulSoup(html, "html.parser") ## 1.웹 브라우저와 연결해서 데이터를 가져오라고 request 하는 건데 webserver한테 우리가 어떤 환경에서 작업하고 있는지를 알려줘야 하는데
##이게 자동으로 가긴 하나 제대로 가지 않을 때가 있다. 그리고 이건 legitamate하지 않은 request라고 말할 때가 있다. 그게 Error 406이다.
```

```python
print(doc.prettify())
```

```python
r_names = doc.find_all(class_="column-2")
```

```python
names = [n.get_text() for n in r_names] ##n.text for n in r_names 해도 된다 
```

```python
r_sales = doc.find_all(class_="column-5")
```

```python
sales = [s.text for s in r_sales]
```

```python
r_ratings = doc.find_all(class_="column-6")
```

```python
ratings = [k.text for k in r_ratings]
```

```python
import pandas as pd
```

```python
pd.DataFrame({"Title" : names,
             "Sales" : sales,
             "Rating" : ratings}) ## 0번째줄에 이미 이름이랑 종류들이 떠 있다
```

```python
pd.DataFrame({"Title" : names[1:],
             "Sales" : sales[1:],
             "Rating" : ratings[1:]}) ##python이 인식하는 1번째 원소부터 끝까지를 의미함.
```

```python
movie_ranking = pd.DataFrame({names[0]: names[1:],
             sales[0] : sales[1:],
             ratings[0] : ratings[1:]})## 꼭 따옴표 하고나서 str 자료로 제목을 지어줄 필요 없이 : 전에 들어가는 요소가 제목으로 들어간다
movie_ranking
```

```python
movie_ranking.head() ## 앞에 5개 가져다 준다
```

```python
movie_ranking.corr() ## 앞에서 각 column에서 text로서 가져왔기 때문에 숫자간의 correlation을 보여주는 corr()을 쓸 수가 없다. corr 쓰고자 하는 자료들을 숫자로 바꿔주어야 한다.
```

```python
for i in sales:
    float(i) ## sales 자료 안에는 $ 표시가 있기 때문에 에러 뜸
```

```python
## $ 제거하기
n_sales = [float(s.replace("$","")) for s in sales[1:]] ## replace("$","")는 $모양을 아무것도 없는 걸로 바꾸라는 말
```

```python
n_ratings = [float(k.replace("%","")) for k in ratings[1:]]
```

```python
movie_ranking = pd.DataFrame({names[0]: names[1:],
             sales[0] : n_sales,
             ratings[0] : n_ratings})
movie_ranking
```

```python
movie_ranking.corr()
```

```python
import numpy as np
np.corrcoef(n_sales, n_ratings) ## 위에 한거 이렇게 한 거 있다
```
