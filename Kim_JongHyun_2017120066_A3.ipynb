```python
import urllib.request as req
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
driver = webdriver.Chrome('/Users/jonghyunkim/repo/Tristan/chromedriver')

url = 'https://www.tripadvisor.com/Restaurants-g294197-zfn15566165-Seoul.html'

driver.get(url)

xpath = ['//*[@id="component_2"]/div/div[1]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[2]/span/div[1]/div[2]/div[1]/div/span/a', 
         '//*[@id="component_2"]/div/div[3]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[4]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[6]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[7]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[8]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[10]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[11]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[12]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[13]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[14]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[15]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[16]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[17]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[18]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[19]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[20]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[21]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[22]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[23]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[24]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[25]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[26]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[27]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[28]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[29]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[30]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[31]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[32]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[33]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[34]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[35]/span/div[1]/div[2]/div[1]/div/span/a',
         '//*[@id="component_2"]/div/div[36]/span/div[1]/div[2]/div[1]/div/span/a'] ##xpath 몇개를 따오다 보니 규칙이 있어서 그냥 수기로 규칙에 따라 바꿔주고 밑에 try로 없는 숫자들 배제
```

```python
a = len(xpath)
Name = []
N_reviews = []
Ratings = []
Address = []
Telephone = []

driver.implicitly_wait(5) ## 코드를 연속으로 실행하고 기다릴 시에 위 코드의 실행이 끝났으나 webdriver는 인터넷 속도에 따라 로딩중이어서 아직 정보가 불러와지지 않았는데
##아래의 코드들을 실행시켜버리면 에러가 발생하는 문제를 해결하기 위한 코드

for n in range(0,a):
    try :
        driver.find_element_by_xpath(xpath[n]).click() ##각 xpath의 규칙에 따라 클릭을 하도록 하고 광고나 규칙 중 빠진 숫자들은 try를 통해 error들은 실행하지 않도록 해준다.
        driver.switch_to.window(driver.window_handles[-1]) ##webdriver를 통해 클릭하니 새탭을 띄우나 그 탭으로 이동은 안하는 현상이 있어 마지막 탭으로 이동시킨다.
        doc = BeautifulSoup(driver.page_source, "html.parser") ## 현재 보여주고 있는 창의 html을 가져온다
        
        name = doc.find(class_="fHibz").text
        Name.append(name)
        
        n_reviews = doc.find(class_="eBTWs").text
        N_reviews.append(n_reviews)
        
        ratings = doc.find(class_="fdsdx").text.strip()
        Ratings.append(ratings)
        
        address = doc.find_all(class_="fhGHT")[1].text
        Address.append(address)
        
        t1 = doc.find_all(class_='dyeJW dUpPX')[2].text ##핸드폰 번호만을 나타내는 좋은 class를 발견하지 못해 강제로 이 class를 가진 것 중 3번째가 telephone이라는 특징을 반영하도록 한다.
        if t1 == "+ Add phone number": ## telephone이 등록되지 않은 경우 '+Add phone number'로 크롤링되는 것을 과제 요구사항대로 'na'로 가져오기 위한 조건문
            Telephone.append('na')
        else:
            Telephone.append(t1)
        driver.close() ##탭을 닫지 않고 계속 진행하면 컴퓨터에 따라서는 ram 용량으로 인해 실행을 멈추는 문제를 위해 현재 탭을 닫고 진행
        driver.implicitly_wait(5) ## 인터넷 환경에 따라 로딩속도 차이로 발생할 수 있는 xpath 실행 문제를 위한 대기
        driver.switch_to.window(driver.window_handles[0]) ## 위에서 탭을 닫아도 첫번째 탭으로 이동을 시켜주는 명령을 하지 않으면 python이 webdriver가 종료되었다고 인식하는 것을 방지
    except :
        print('xpath', n, 'is not a path') ##위의 xpath 특징에 따라 만든 list에서 특징의 예외들로 인해 발생하는 오류들을 무시하고 구분해서 몇번째 원소인지 알려달라는 구문
    
driver.quit() ##작업이 끝났으나 webdriver는 닫아준다.
    
        
print(Name, len(Name))
print(N_reviews, len(N_reviews))
print(Ratings, len(Ratings))
print(Address, len(Address))
print(Telephone, len(Telephone)) ## 뒤에 붙인 각 list의 length는 pandas 사용시 각 list의 원소 개수 차이로 인한 error 발생방지. + 각 list에 잘 append 되었는지 확인.
```
